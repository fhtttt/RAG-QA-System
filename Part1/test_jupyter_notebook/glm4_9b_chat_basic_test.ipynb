```python
from openai import OpenAI

base_url = "http://127.0.0.1:9091/v1/"
client = OpenAI(api_key="EMPTY", base_url=base_url)
```

```python
messages = [{"role": "user", "content": "Hi Please introduce yourself"}]
```

```python
response = client.chat.completions.create(
    model="glm-4",
    messages=messages,
    max_tokens=1024,
    temperature=0.8, 
)
```
